## 이상치 제거

# Feature Extraction

### 1)Age_Grouping

연속적인 수치 데이터 

->
 범위를 지정하거나 카테고리화를 통해 이전보다 작은 수의 그룹으로 만듦 

```python
df.loc[df['Age']<= 15,'Age_group']=0
df.loc[(df['Age']>15) & (df['Age']<=30),'Age_group']=1
df.loc[(df['Age']>30) & (df['Age']<=45),'Age_group']=2
df.loc[(df['Age']>45) & (df['Age']<=60),'Age_group']=3
df.loc[df['Age'] > 60,'Age_group']=4
```

```python
df['Age_group']=df['Age_group'].astype('int')
```

```python
df
```

**Feature 추출이라고도 할 수 있음 !**

## SibSp + Parch

형제,자매,배우자,부모님,자녀 수를 합한 ***Family_size***라는 새로운 Feature 추출

```python
df["Family_size"]=df["SibSp"]+df['Parch']+1
df['Family_size']=df['Family_size'].astype(int)
```

# 머신러닝 Training

```python
df_X=df_X.values
df_y=df_y.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_X,df_y,test_size=0.3,random_state=0)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
knn=KNeighborsClassifier()
svm=SVC()
rf=RandomForestClassifier()
```

```python
knn.fit(X_train,y_train)
svm.fit(X_train,y_train)
rf.fit(X_train,y_train)
```

# 머신러닝 Test

```python
pred_data=knn.predict(X_test)
pred_data2=svm.predict(X_test)
pred_data3=rf.predict(X_test)
```

```python
pred_data
```

결과 분석 -정확도 확인

```python
pred_data = pred_data.reshape(-1,1)
pred_data2 = pred_data2.reshape(-1,1)
pred_data3 = pred_data3.reshape(-1,1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,pred_data))
print(accuracy_score(y_test,pred_data2))
print(accuracy_score(y_test,pred_data3))
```
